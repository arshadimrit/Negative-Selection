```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as graph
```

```python
df = pd.read_csv('/home/arshad/Desktop/caste-genes.csv')
df.head()
```

```python
select_columns = ['protein_id', 'accession number', 'description', 'sequence', 'd_ave', 'q_ave', 'w_ave', 'dvw', 'dvq', 
                  'qvw', 'Ave_fold_d', 'Ave_fold_q']
```

```python
# The Honey Bee Protein Atlas (Chan et al. 2013) provides protein expression
# data in 26 tissues in queens and workers for 1,728 proteins in OGSv3.2. We examined if
# significantly worker-biased proteins, averaged across the different tissues (Chan et al. 2013),
# have different γ relative to significantly queen-biased proteins using a Wilcoxon non-parametric
# test. We also counted the number of cases where worker-biased proteins were enriched for
# nonsynonymous outlier SNPs relative to all proteins found in a given tissue; this analysis was
# repeated for 26 tissues and for each of the 6 pairwise population comparisons (a total of 156
# tests). We performed a similar analysis for queen-biased genes. After first ensuring that queen biased
# and worker-biased proteins did not significantly differ in length, we compared the number
# of significant and non-significant (FDR; α<0.05) tests of enrichment in worker-biased and
# queen-biased proteins using a Fisher’s Exact test. The Bee Protein Atlas also provided a
# proteomic contrast of drones and workers. Worker-biased proteins had higher selection
# coefficients relative to drone-biased proteins but the number of drone-biased proteins was too
# small to warrant a statistical analysis. used the qvw column specifically

# Wilcoxon non-parametric test = Drone vs queen, different γ
# Fisher's Exact test = Enrichment (FDR; α<0.05)
```

```python
# "quant_d", "quant_q", "quant_w" - levels of a given protein, in a given organ, expressed in percentage (d+q+w=100%)
# "d_colour","q_colour","w_colour" - protein percentages expressed in grayscale
# "dsigB","qsigB" - values with p<0.05 are indicated by an asterisk, drone or queen relative to worker levels
# "d_ave","q_ave","w_ave" - whole body average for a given protein
# "dvw","dvq","qvw" - whole body average with p<0.05 are indicated by an asterisk
# "fold_d","fold_q" - fold-difference of a given protein, in a given organ, in drone or queen relative to worker levels
# "Ave_fold_d","Ave_fold_q" - fold-difference of a given protein in drone or queen relative to worker levels averaged across the whole body
```

```python
caste = df[select_columns]
caste['genes'] = caste['description'].str.split(':').str[0].str.split('PREDICTED').str[0]
caste['ncbi_format_genes'] = caste['genes'].str.split('.').str[0].str.split('_').str[1]
caste.drop(caste.tail(14).index, inplace=True)
caste.replace('\\N', np.nan)
caste.head()
```

```python
significant_drone_queen = caste.loc[caste['dvq'] == '*']
```

```python
significant_drone_queen.head()
```

```python
object_columns = ['d_ave', 'q_ave', 'w_ave', 'Ave_fold_d', 'Ave_fold_q']
for column in object_columns:
    significant_drone_queen[column] = significant_drone_queen[column].str.replace('\\\\N', 'NaN').astype(float)
```

```python
significant_drone_queen['dronevsqueenfold'] = significant_drone_queen['Ave_fold_d'] / significant_drone_queen['Ave_fold_q']
```
