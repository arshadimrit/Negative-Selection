```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as graph
from tqdm import tqdm
from Bio import SeqIO

start_codon = 'ATG'
synonymous_list = ['CTT', 'CTC', 'CTA', 'CTG', 
                  'GTT', 'GTA', 'GTC', 'GTG', 
                  'GCT', 'GCC', 'GCA', 'GCG', 
                  'GGT', 'GGC', 'GGA', 'GGG', 
                  'CCT', 'CCA', 'CCG', 'CCC',
                  'ACT', 'ACG', 'ACC', 'ACA',
                  'TCT', 'TCC', 'TCA', 'TCG',
                  'CGT', 'CGC', 'CGA', 'CGT']

lowercase_synonymous_list = ['ctT', 'ctC', 'ctA', 'ctG', 
                  'gtT', 'gtA', 'gtC', 'gtG', 
                  'gcT', 'gcC', 'gcA', 'gcG', 
                  'ggT', 'ggC', 'ggA', 'ggG', 
                  'ccT', 'ccA', 'ccG', 'ccC',
                  'acT', 'acG', 'acC', 'acA',
                  'tcT', 'tcC', 'tcA', 'tcG',
                  'cgT', 'cgC', 'cgA', 'cgT']

input_file = '/home/arshad/Downloads/am45new.fasta'
```

```python
df = pd.DataFrame(columns=['ID','seq'])
```

```python
fasta_sequences = SeqIO.parse(open(input_file),'fasta')

fasta_id = []
fasta_seq = []

for fasta in tqdm(fasta_sequences):
    name, sequence = fasta.id, str(fasta.seq)
    fasta_id.append(name)
    fasta_seq.append(sequence)
```

```python
matches = []
total = 0
for scaffold in tqdm(fasta_seq):
    same = [codon_syn for codon_syn in synonymous_list if codon_syn in scaffold]
    matches.append(same)
    total = total + len(same)
print(total)
```
