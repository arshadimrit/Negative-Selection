```python
sns.distplot(liquid_data['plank'], label='PLANK')
sns.distplot(liquid_data['run'], label='RUN')
sns.distplot(liquid_data['erg'], label='ERG')
graph.legend()
graph.show()

sns.regplot(liquid_data['plank'], liquid_data['run'])
graph.show()
```

```python
x_plank = liquid_data['plank'].values.reshape(-1, 1)
x_erg = liquid_data['erg'].values.reshape(-1, 1)
y_run = liquid_data['run'].values.reshape(-1, 1)
```

```python
linear_liquid_run_erg = LinearRegression().fit(x_erg, y_run)
linear_liquid_run_plank = LinearRegression().fit(x_plank, y_run)

print(f'Run and erg: coeff= {linear_liquid_run_erg.coef_}')
print(f'Run and plank: coeff= {linear_liquid_run_plank.coef_}')
```

```python
run_pred_erg = linear_liquid_run_erg.predict(x_erg)
run_pred_plank = linear_liquid_run_erg.predict(x_plank)


graph.title('ERG and PLANK')
sns.regplot(x_erg.flatten(), x_plank.flatten())
graph.show()

graph.title('Run and ERG')
sns.regplot(y_run, run_pred_erg.flatten())
graph.show()

graph.title('Run and PLANK')
sns.regplot(y_run, run_pred_plank.flatten())
graph.show()
```
